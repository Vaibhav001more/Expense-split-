```python
df['decision'].value_counts().plot(kind='bar', title='Toss Decisions')
plt.ylabel('Count')
plt.show()
```

```python
# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Create a new column 'season' from year
df['season'] = df['date'].dt.year

# Plot the number of matches per season
df['season'].value_counts().sort_index().plot(kind='bar', title='Matches per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
toss_win = df[df['toss_winner'] == df['winner']]
percentage = (len(toss_win) / len(df)) * 100
print(f"Winning Toss and Match: {percentage:.2f}%")
```
